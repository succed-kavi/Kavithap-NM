```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report

from google.colab import files
uploaded=files.upload()
true_df = pd.read_csv('True.csv')
true_df['label'] = 0 # Add a label column for true news

```

```python
true_df.head()
```

```python
from google.colab import files
uploaded=files.upload()
fake_df = pd.read_csv('Fake.csv')
fake_df['label'] = 1
```

```python
fake_df.head()
```

```python
df = pd.concat([true_df, fake_df], ignore_index=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
```

```python
X_test
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
accuracy = accuracy_score(y_test, predictions)
classification_report_result = classification_report(y_test, predictions)

```

```python
accuracy
```

```python
print(f'Accuracy: {accuracy}')
print('Classification Report:\n', classification_report_result)
```

```python
user_input = input("Enter a news article: ")

# Make a prediction
prediction = model.predict([user_input])

# Display the result
if prediction[0] == 0:
    print("The news is likely to be true.")
else:
    print("The news is likely to be fake.")
```

```python
import gradio as gr

def predict_fake_news(text):
  """Predicts whether a news article is fake or real.

  Args:
    text: The text of the news article.

  Returns:
    A string indicating whether the news is likely to be true or fake.
  """
  prediction = model.predict([text])
  return "The news is likely to be true." if prediction[0] == 0 else "The news is likely to be fake."

interface = gr.Interface(
    fn=predict_fake_news, # Now this function is defined
    inputs=gr.Textbox(label="Enter News Article Text"),
    outputs=gr.Text(label="Prediction"),
    title="📰 Fake News Detector",
    description="Enter a news headline or article text to check if it's likely fake or real.",
    theme="soft"
)

interface.launch()
```
